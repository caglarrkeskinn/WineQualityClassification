```python
data = pd.read_csv("winequality-red.csv")
data
```

```python
x_data = data[['fixed acidity','volatile acidity','citric acid','residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide','density','pH','sulphates','alcohol']]
y_data = data[['quality']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x_data, y_data, test_size=0.3, random_state=1)
```

```python
# DECISION TREE CLASSIFIER
dt = DecisionTreeClassifier(random_state=0)
# train the model
dt.fit(x_train, y_train)
# make predictions
dt_pred = dt.predict(x_test)
# print the accuracy
print("Accuracy of Decision Tree Classifier: ",
      accuracy_score(y_test, dt_pred))
```

```python
target_names = ['class 0', 'class 1', 'class 2']
print(classification_report(y_data, dt_pred, target_names=target_names))




```
